<p style="text-align: center;"><em>Feature engineering - Outliers</em></p>

```python
df.isnull().sum()
```

```python
df = df.loc[:, (df.isnull().mean() * 100 <= 10) | (df.columns == 'SalePrice')]
```

```python
nulldf=df[df.columns[df.isnull().sum() > 0]]
```

```python
nulldf.head()
```

```python
def fill_missing_values(df, group_col, fill_col, method='mean'):
    if method == 'mean':
        df[fill_col] = df.groupby(group_col)[fill_col].transform(lambda x: x.fillna(x.mean()))
    elif method == 'mode':
        df[fill_col] = df.groupby(group_col)[fill_col].transform(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else x))

fill_params = [
    ('Neighborhood', 'MSZoning', 'mode'),
    ('Neighborhood', 'Utilities', 'mode'),
    ('Neighborhood', 'Exterior1st', 'mode'),
    ('Neighborhood', 'Exterior2nd', 'mode'),
    ('Neighborhood', 'MasVnrArea', 'mean'),
    ('Neighborhood', 'BsmtQual', 'mode'),
    ('Neighborhood', 'BsmtCond', 'mode'),
    ('Neighborhood', 'BsmtExposure', 'mode'),
    ('Neighborhood', 'BsmtFinType1', 'mode'),
    ('Neighborhood', 'BsmtFinType2', 'mode'),
    ('Neighborhood', 'Electrical', 'mode'),
    ('Neighborhood', 'KitchenQual', 'mode'),
    ('Neighborhood', 'Functional', 'mode'),
    ('Neighborhood', 'SaleType', 'mode'),
    ('Neighborhood', 'GarageType', 'mode'),  
    ('Neighborhood', 'GarageFinish', 'mode'),
    ('GarageType', 'GarageCars', 'mean'),
    ('GarageType', 'GarageArea', 'mean'),
    ('GarageType', 'GarageQual', 'mode'),
    ('GarageType', 'GarageCond', 'mode'),
    ('Neighborhood', 'BsmtFinSF1', 'mean'),
    ('Neighborhood', 'BsmtFinSF2', 'mean'),
    ('Neighborhood', 'BsmtUnfSF', 'mean'),
    ('Neighborhood', 'TotalBsmtSF', 'mean'),
    ('Neighborhood', 'BsmtFullBath', 'mode'),
    ('Neighborhood', 'BsmtHalfBath', 'mode')
]

for group_col, fill_col, method in fill_params:
    fill_missing_values(df, group_col, fill_col, method)
```

```python
df['TotalSF'] = df['TotalBsmtSF'] + df['1stFlrSF'] + df['2ndFlrSF']
df['TotalBathrooms'] = df['FullBath'] + df['BsmtFullBath'] + 0.5 * (df['HalfBath'] + df['BsmtHalfBath'])
df['HouseAge'] = df['YrSold'] - df['YearBuilt']
df['RemodAge'] = df['YrSold'] - df['YearRemodAdd']
df['TotalRooms'] = df['TotRmsAbvGrd'] + df['TotalBsmtSF']
df['OverallGrade'] = df['OverallQual'] * df['OverallCond']
df['GarageScore'] = df['GarageArea'] * df['GarageCars']
```

```python
df['GarageYrBlt'].fillna(df['YearBuilt'], inplace=True)
```

```python
df = df.drop(['PoolArea', 'MSSubClass', 'OverallCond', 'MoSold', '3SsnPorch', 'YrSold', 'LowQualFinSF', 'Id', 'MiscVal', 'BsmtHalfBath', 'BsmtFinSF2'],axis=1)
```

<p style="text-align: center;"><em>Modelling</em></p>

```python
df=pd.get_dummies(df,drop_first=True)
```

```python
train= df[df['SalePrice'].notna()]
test= df[df['SalePrice'].isna()]
```

```python
x=train.drop("SalePrice",axis=1)
y=train["SalePrice"]
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge,Lasso
from sklearn.linear_model import ElasticNet
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.tree import ExtraTreeRegressor
from sklearn.tree import DecisionTreeRegressor
from xgboost import XGBRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error


def algo_test(x,y):

        L=LinearRegression()
        R=Ridge()
        Lass=Lasso()
        E=ElasticNet()
        ETR=ExtraTreeRegressor()
        GBR=GradientBoostingRegressor()
        kn=KNeighborsRegressor()
        dt=DecisionTreeRegressor()
        xgb=XGBRegressor()
       
        
        algos=[L,R,Lass,E,ETR,GBR,kn,dt,xgb]
        algo_names=['Linear','Ridge','Lasso','ElasticNet','Extra Tree','Gradient Boosting','KNeighborsRegressor','Decision Tree','XGBRegressor']
        
        x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=.2,random_state=42)
        
        r_squared= []
        rmse= []
        mae= []
        
        result=pd.DataFrame(columns=['R_Squared','RMSE','MAE'],index=algo_names)
        
        
        for algo in algos:
            p=algo.fit(x_train,y_train).predict(x_test)
            r_squared.append(r2_score(y_test,p))
            rmse.append(mean_squared_error(y_test,p)**.5)
            mae.append(mean_absolute_error(y_test,p))
        
        result.R_Squared=r_squared
        result.RMSE=rmse
        result.MAE=mae
        
        rtable=result.sort_values('R_Squared',ascending=False)
        return rtable
```

```python
print("Best Model:", algo_test(x, y))
```
